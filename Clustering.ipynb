Q1.What is unsupervised learning in the context of machine learning?

Ans:-

Unsupervised learning is a type of machine learning where the model learns patterns from data without using labeled outputs. It finds hidden structures in the dataset automatically.

Example: Customer segmentation

Q2.How does K-Means clustering algorithm work?

Ans:-

K-Means clustering divides the dataset into K clusters. It selects random centroids, assigns data points to the nearest centroid, updates the centroid, and repeats the process until clusters stop changing.

---
Q3.Explain the concept of a dendrogram in hierarchical clustering.

Ans:-

A dendrogram is a tree-like diagram used in hierarchical clustering. It shows how clusters are formed step-by-step and how similar the data points are.

----
Q4.What is the main difference between K-Means and Hierarchical Clustering?

Ans:-

K-Means divides data into a fixed number of clusters, while hierarchical clustering builds clusters step-by-step without fixing the number of clusters in advance.

----
Q5.What are the advantages of DBSCAN over K-Means?

Ans:-

DBSCAN can detect clusters of different shapes and can also identify noise points. K-Means cannot detect noise and works best only with spherical clusters.

----
Q6.When would you use Silhouette Score in clustering?

Ans:-

Silhouette Score is used to evaluate how well the data points are grouped into clusters. It helps in selecting the best number of clusters.

----
Q7.What are the limitations of Hierarchical Clustering?

Ans:-

Hierarchical clustering is slow for large datasets and requires more memory. Once a cluster is formed, it cannot be changed later.

----
Q8.Why is feature scaling important in clustering algorithms like K-Means?

Ans:-

Feature scaling is important because K-Means uses distance to form clusters. If features are not scaled, one feature may dominate the clustering process.

----
Q9.How does DBSCAN identify noise points?

Ans:-

DBSCAN identifies noise points as those data points that do not belong to any dense region. These points are far from clusters and are marked as outliers.

---
Q10.Define inertia in the context of K-Means.

Ans:-

Inertia is the sum of squared distances between data points and their cluster centroids. Lower inertia means better clustering.

---
Q11.What is the elbow method in K-Means clustering?

Ans:-

The elbow method is used to find the best number of clusters. It plots inertia values for different K values and selects the point where the curve bends.

----
Q12.Describe the concept of density in DBSCAN.

Ans:-

Density in DBSCAN means the number of points present in a region. A cluster is formed when many points are close to each other.

---
Q13.Can hierarchical clustering be used on categorical data?

Ans:-

Yes, hierarchical clustering can be used on categorical data if an appropriate distance measure is used.

----
Q14.What does a negative Silhouette Score indicate?

Ans:-

A negative Silhouette Score indicates that the data points are assigned to the wrong clusters.

----
Q15.Explain the term linkage criteria in hierarchical clustering.

Ans:-

Linkage criteria define how the distance between clusters is calculated. Examples include single linkage, complete linkage, and average linkage.

----
Q16.Why might K-Means perform poorly on data with varying cluster sizes?

Ans:-

K-Means assumes all clusters have similar sizes and shapes. If clusters are very different, the algorithm may give incorrect results.

----
Q17.What are the core parameters in DBSCAN?

Ans:-

The main parameters in DBSCAN are eps (distance value) and min_samples (minimum points needed to form a cluster). These control how clusters are formed.

----
Q18.How does K-Means++ improve standard K-Means?

Ans:-

K-Means++ selects better initial centroids instead of random ones. This improves clustering accuracy and reduces errors.

----
Q19.What is agglomerative clustering?

Ans:-

Agglomerative clustering is a type of hierarchical clustering where each point starts as its own cluster and clusters are merged step-by-step.

----
Q20.What makes Silhouette Score better than inertia?

Ans:-

Silhouette Score considers both cluster distance and cluster separation, while inertia only measures distance inside clusters.

Q21.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("K-Means Clustering")
plt.show()
```

Q29.Create synthetic data using make_blobs and evaluate silhouette scores for  k = 2 to 5. Display as a bar chart.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4, random_state=42)

scores = []

for k in range(2,6):
    model = KMeans(n_clusters=k, random_state=42)
    labels = model.fit_predict(X)
    scores.append(silhouette_score(X, labels))

plt.bar(range(2,6), scores)
plt.title("Silhouette Score")
plt.show()
```

Q34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot

```python
from sklearn.datasets import *
from sklearn.cluster import *
from sklearn.preprocessing import *
from sklearn.decomposition import *
from sklearn.metrics import *
from sklearn.manifold import *
import matplotlib.pyplot as plt
import numpy as np
data = load_breast_cancer()
X = data.data

inertia = []

for k in range(2,7):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    inertia.append(model.inertia_)

plt.plot(range(2,7), inertia)
plt.title("Inertia vs K")
plt.show()
```

Q40.Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D
scatter plot
